```python
# Import of libraries and dataset
import pandas as pd
import numpy as np

df= pd.read_csv("/Users/EB/Desktop/NU/ANA500/wk1/StudentsPerformance_with_header.csv")

df.head()
```

```python
# Sort datase by Grade
df.sort_values(by='GRADE', ascending=False).head()
```

```python
#Check for missing
df.isnull().sum()
```

```python
# Check for duplicate
df.duplicated().sum()
```

```python
# Calculation Normalize Grade values
df['GRADE_normalized'] = df['GRADE'].apply(lambda x: round(x / df['GRADE'].max(), 2))
df[['GRADE', 'GRADE_normalized']]
```

```python
# Statistical analysis
df[['Additional work', 'Graduated high-school type', 'Scholarship type','Attendance to classes', 'Weekly study hours', 'Preparation to midterm exams 1', 'Preparation to midterm exams 2','Impact of your projects/activities on your success', 'GRADE']].describe()
```

```python
# Correlation Heatmap showing variables 
import seaborn as sns
import matplotlib.pyplot as plt

df_sel = df[
    [
        "Additional work",
        "Graduated high-school type",
        "Scholarship type",
        "Attendance to classes",
        "Weekly study hours",
        "Preparation to midterm exams 1",
        "Preparation to midterm exams 2",
        "Impact of your projects/activities on your success",
        "GRADE",
    ]]

plt.figure(figsize=(10,6))
sns.heatmap(df_sel.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
#Saving Dataset

df.to_csv("cleaned_student_performance.csv", index=False)
```
