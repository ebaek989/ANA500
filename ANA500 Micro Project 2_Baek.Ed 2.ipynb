```python
# Import of libraries and dataset
import pandas as pd
import numpy as np

df= pd.read_csv("/Users/EB/Desktop/NU/ANA500/employee attrition.csv")

df.head()
```

```python
#Check for missing
df.isnull().sum()
```

```python
# Check for duplicate
df.duplicated().sum()
```

```python
df["Attrition New"] = df["Attrition"].replace({"No": 0, "Yes": 1}).astype("Int64")

df["Gender New"] = df["Gender"].replace({"Female": 1, "Male": 0}).astype("Int64")

df["BusinessTravel New"] = (
    df["BusinessTravel"]
      .replace({"Non-Travel": 0, "Travel_Rarely": 1, "Travel_Frequently": 2})
      .astype("Int64"))

df["OverTime New"] = df["OverTime"].replace({"No": 0, "Yes": 1}).astype("Int64")
```

```python
df.describe()
```

```python
# Correlation Heatmap showing variables 
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
#Saving Dataset

df.to_csv("cleaned_attrition.csv", index=False)

cleaned_df=pd.read_csv("cleaned_attrition.csv")

cleaned_df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

df = pd.read_csv("/Users/EB/Desktop/NU/ANA500/wk2/cleaned_attrition.csv")
```

```python
#Heatmap
num = df.select_dtypes(include="number")
corr = num.corr()
target = "Attrition New"
col = corr[[target]].sort_values(by=target, ascending=False)

plt.figure(figsize=(10,6))
sns.heatmap(col, annot=True, cmap="coolwarm")
plt.title("Correlation of Variables with Attrition")
plt.tight_layout(); plt.show()
```

```python
# Boxplot Age by Attrition
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Attrition", y="Age")
plt.title("Age Distribution by Attrition")
plt.xlabel("Attrition"); plt.ylabel("Age")
plt.tight_layout(); plt.show()
```

```python
# Boxplot Years at Company by Attrition
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Attrition", y="YearsAtCompany")
plt.title("Years at Company by Attrition")
plt.xlabel("Attrition"); plt.ylabel("Years at Company")
plt.tight_layout(); plt.show()
```

```python
# Histogram Monthly Income by Attrition
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="MonthlyIncome", hue="Attrition")
plt.title("Monthly Income Distribution by Attrition")
plt.xlabel("Monthly Income")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Bar Chart Overtime By Attrition
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="OverTime", hue="Attrition")
plt.title("Attrition by OverTime Status")
plt.xlabel("OverTime")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
# Bubble Chart Business Trabel by Attrition
bubble_df = df.groupby(["BusinessTravel", "Attrition"]).size().reset_index(name="Count")

fig = px.scatter(bubble_df, 
                 x="BusinessTravel", 
                 y="Attrition", 
                 size="Count", 
                 color="Attrition",
                 title="Attrition by Business Travel (Bubble Chart)",
                 labels={"BusinessTravel":"Business Travel", "Attrition":"Attrition"},
                 size_max=60)

fig.show()
```
